## Observations from results:

1. From combos 1 & 2 we see that using minimum temperature does not reduce prediction error. Moreover, combo 5, which uses average temperature and average wind speed, has higher error in comparison with combo 1 which uses maximum temperature and average wind speed. This implies that from temperature measurements, maximum temperature has the highest effect on the predicted reference evapotranspiration. This is intuitive because high temperatures result in more evaporation and therefore higher evapotranspiration. Therefore, the maximum temperature will have a greater effect on the total measured evapotranspiration than minimum or average temperature.

2. Using extraterrestrial radiation, which can be computed and does not require field measurement, results in significant improvement in ET0 prediction accuracy when using MLR (see combos 7 & 8). However, we see that this addition resulted in a significant improvement when using MLR whereas no improvement can be seen when using SVR. This can be attributed to the non-linearity of the SVR model. Extraterrestrial radiation is a non-linear function of month (or day of the year) and latitude. Month is an input feature for both combinations 7 & 8, and latitude does not vary significantly between the studied stations.

Plotting month and extraterrestrial radiation

```python
x = df['month']
y = df['Ra']

plt.figure(figsize=(8, 6))
plt.scatter(x=x, y=y, color='limegreen', edgecolor='w', lw=0.5)
plt.xlabel('Month')
plt.ylabel('Extraterrestrial radiation')
plt.title('Month and Extraterrestrial Radiation')
plt.show()
```

3. Comparing combos 7 & 10 and combos 12 & 14, we see that adding average humidity had minimal contribution to improving ET0 predictions.

4. Although maximum temperature proved to be a good predictor of reference evapotranspiration, we notice that in combinations 11 & 12, adding maximum temperature resulted in minimal reduction in prediction error. This can be attributed to the fact that both temperature and solar radiation contribute to the energy part of evapotranspiration causation, the other being surrounding air humidity. Therefore, when solar radiation value is present, adding maximum temperature does not add a lot of new information.

Plotting maximum temperature and solar radiation

```python
x = df['max_temp']
y = df['inc_rad']

plt.figure(figsize=(8, 6))
plt.scatter(x=x, y=y, color='limegreen', edgecolor='w', lw=0.5)
plt.xlabel('Maximum temperature')
plt.ylabel('Solar radiation')
plt.title('Maximum temperature and Solar radiation')
plt.show()
```

5. There is a large gap between train and test errors in several combinations (3, 6, 11, 12, 14), whereby the training dataset has a larger error. Apart from combination No. 6, all the combinations having a large gap do not include wind speed as an input feature. Moreover, comparing combos 11 & 13, we see that adding the average wind speed in combination 13 resulted in a great improvement in training error, but almost no change in test error. **(EXPLORE TRAIN DATASET STATION ERRORS)**

Exploring SVR combinations 11 and 13

```python
svr11 = svr_models.get_model(11)
svr13 = svr_models.get_model(13)
```

```python
st_scores11 = svr11.get_station_scores()
st_scores13 = svr13.get_station_scores()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 5))

metric = 'RMSE'
graph = st_scores11[['st_num', metric]].merge(st_scores13[['st_num', metric]], on='st_num', suffixes=['_11', '_13'])
graph = graph.sort_values(by='st_num')

ax1 = axs[0]
graph.set_index('st_num').plot(
    kind='bar',
    color=['skyblue', 'limegreen'],
    ax=ax1,
    edgecolor='w',
    alpha=0.8)
ax1.set_ylabel('RMSE')
ax1.set_xlabel('Station number')
ax1.set_title('Combinations 11 and 13')

ax2 = axs[1]
graph = df.groupby('st_num').mean()[['avg_ws']]
graph.plot(kind='bar', ax=ax2, color='darkorange', alpha=0.7)
ax2.set_xlabel('Station number')
ax2.set_ylabel('Average wind speed')
ax2.set_title('Average wind speed per station')

plt.show()
```

A large difference in error can be seen in stations 17227 and 17824. These two stations have a significantly higher wind speed average than other stations. This explains the huge drop in error when using average wind speed as an input feature.

```python
st_num = 17227

ax1 = svr11.plot_station_series(which='train', st_num=st_num)
title = 'ET0 Measurements vs Predictions / Combination: 11 / Station: {}'.format(st_num)
ax1.set_title(title)
plt.show()

ax2 = svr13.plot_station_series(which='train', st_num=st_num)
title = 'ET0 Measurements vs Predictions / Combination: 13 / Station: {}'.format(st_num)
ax2.set_title(title)
plt.show()
```
