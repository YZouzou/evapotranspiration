# Splitting data
In this notebook, the data is split into train and test datasets. 

The practical use of the models is to have them trained on regional data from weather stations and then use them to predict reference evapotranspiration in areas with limited climatic variables. Therefore, to simulate the real-life application, test data should come from areas other than train data. This renders random test-train split inapplicable for this study case. The data must be split so that we have stations for training and different stations for testing the model.

## Functions

```python
def split_data(data, test_ratio = 0.4):
    
    '''
    This function splits the data to train and test datasets so that train stations are different from test stations.
    
    Returns a tuple: (train, test)
    
    The function works as follows:
        1. A station is randomly sampled from the data and added to the list of test stations
        2. The number of instances in the stations are computed and added to the total number
           of instances of the test stations
        3. Once the total number of instances in test data exceeds the desired value (test_ratio*total_instances)
           the function returns the train and test datasets.
           
    Note that the test ratio will not be exact because splitting is done on station basis rather than splitting
    instances randomly from all stations.
    
    '''
    test_size = test_ratio * data.shape[0]
    
    # Series of unique stations in the dataset
    stations = pd.Series(data['st_num'].unique())
    
    st_list = []
    count = 0
    
    while count < test_size:
        st = stations.sample()
        st_list.append(st.values[0])
        stations = stations.drop(index = st.index)
        
        cond = data['st_num'] == st.values[0]
        count += cond.sum()
    
    cond = data['st_num'].isin(st_list)
    test = data[cond].reset_index(drop = True)
    train = data[~cond].reset_index(drop = True)

    print('Total dataset size: {}'.format(data.shape[0]))
    print('Test dataset size: {}'.format(test.shape[0]))
    print('Test dataset ratio: {:.3f}'.format(float(test.shape[0]/data.shape[0])))
    print('\n')
    print('Total number of stations: {}'.format(data['st_num'].unique().shape[0]))
    print('Number of stations in test dataset: {}'.format(test['st_num'].unique().shape[0]))

    return (train, test)
```

```python
def plot_map(stations,
             color='blue',
             path_to_st_def = r"C:\Users\HP\Desktop\Git\evapotranspiration\processed_data\station_definitions.csv"):
    
    '''
    Plot stations in data with pop-ups for st_num.
    If color == 'color', a column named 'color' is expected to contain the color of each station.
    Else, all stations will have the same color defined in color parameter.
    
    Parameters:
    -----------
    
    stations: list
        A list of station numbers to be plotted.
        
    color: str, dict
        Color to use for stations.
        Could be a color code to be used for all stations or a dictionary mapping station numbers to colors.
        
    path_to_st_def: str
        Path to the station_definitions.csv file containing station numbers, latitudes, and longitudes.
        
    '''
    
    tiles = 'https://stamen-tiles-{s}.a.ssl.fastly.net/terrain-background/{z}/{x}/{y}{r}.png'
    attr = 'Map tiles by <a href="http://stamen.com">Stamen Design</a>, <a href="http://creativecommons.org/licenses/by/3.0">CC BY 3.0</a> &mdash; Map data &copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors'
    location = [37.5, 28.5]
    zoom = 8
    marker_radius = 4
    
    # Loading station definitions
    st_def = pd.read_csv(path_to_st_def)[['st_num', 'latitude', 'longitude', 'elevation']]

    # Merging station definitions with station scores
    cond = st_def['st_num'].isin(stations)
    df = st_def.loc[cond].reset_index(drop=True)
    
    m = folium.Map(location = location, zoom_start = zoom, tiles = tiles, attr = attr)
    
    # Assigning the color column'
    if isinstance(color, dict):
        df['color'] = df['st_num'].replace(color)
    else:
        df['color'] = color

    for row in df.iterrows():
        row = row[1]

        # Location of station
        loc = [row['latitude'], row['longitude']]

        # Text in pop-up
        txt = ''
        txt += '<h5><b>Station: {}\n</h5></b>'.format(int(row['st_num']))
        txt += 'Elevation: {} m'.format(row['elevation'])
        popup = folium.Popup(txt, min_width = 100, max_width = 100)
        folium.CircleMarker(location = loc,
                            fill = True,
                            fill_opacity = 0.8,
                            opacity = 1,
                            color = row['color'],
                            radius = marker_radius
                            ).add_child(popup).add_to(m)

    display(m)

    return m
```

```python
def train_test_dist(ax, train_data, test_data, var, bins, title, alpha, colors):
    
    '''
    Plot the histograms of train, test, and combined data of the variable defined in var.
    The histograms are plotted on the same axes object and behind each other (unstacked).
    
    Parameters:
    -----------
    
    ax: matplotlib.axes
    
    train_data, test_data: pandas.DataFrame
        Dataframes of train and test data containing column defined in var
        
    var: str
        Name of the column containing the data to be plotted
        
    bins: int
        Number of bins of the histogram
        
    title: str
    
    alpha: float (0-1)
        Opacity of the bars
    
    colors: list
        List of three colors in the following order:
            [all_data, train_data, test_data]
            
    '''
    all_data = pd.concat([train_data, test_data])
    count, edges = np.histogram(all_data[var], bins = bins)
    labels = list(map(lambda x: '{:.1f}'.format(x), edges))
    
    # All Data
    ax.hist(all_data[var], bins = edges, edgecolor = 'w', alpha = alpha, color = colors[0], label = 'All')
    
    # Training Data
    ax.hist(train_data[var], bins = edges, edgecolor = 'w', alpha = alpha, color = colors[1], label = 'Training')
    
    # Test Data
    ax.hist(test_data[var], bins = edges, edgecolor = 'w', alpha = alpha, color = colors[2], label = 'Test')
    
    ax.set_xticks(ticks = edges)
    ax.set_xticklabels(labels, rotation = 'vertical')
    ax.set_title(title)
    ax.legend()
    
    
def plot_split_distributions(train, test, figsize = (14, 14)):
    
    '''
    Plot the distributions of the variables defined in var_list inside the function definition.
    Uses train_test_dist() function to plot the distribution of each variable.
    
    '''
    
    var_list = ['min_temp', 'max_temp', 'avg_hum', 'avg_ws', 'inc_rad', 'ET0']
    
    fig, axs = plt.subplots(3, 2, constrained_layout = True, figsize = (14, 14))
    
    fig.suptitle('Training and Test Data Distributions\n', fontsize = 20)
    axs = axs.flatten()
    
    
    bins = 25
    titles = ['Minimum Temperature', 'Maximum Temperature', 'Maximum Humidity',
              'Wind Speed', 'Solar Radiation', 'FAO-56 PM ET0']

    alpha = 0.6
    colors = ['grey', 'black', '#1c6fff']
    for i, var in enumerate(var_list):
        train_test_dist(axs[i], train, test, var, bins, titles[i], alpha, colors)

    plt.show()
```

```python
def print_split_details(train,
                        test,
                        features = ['year', 'elevation', 'max_temp', 'min_temp', 'avg_temp',
                                    'avg_ws', 'max_hum', 'min_hum', 'avg_hum', 'inc_rad', 'Ra', 'ET0']):
    
    data_prop = pd.DataFrame(columns = ['feature', 'train_test', 'min', 'max', 'mean', 'std']).set_index(['feature', 'train_test'])

    function_dict = dict(zip(
    ['min', 'max', 'mean', 'std'],
    [np.min, np.max, np.mean, np.std]))

    for feature in features:
        for col in data_prop.columns:
            data_prop.loc[(feature, 'train'), col] = function_dict[col](train[feature])
            data_prop.loc[(feature, 'test'), col] = function_dict[col](test[feature])
    return data_prop
```
