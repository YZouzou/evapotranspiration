## Training models

```python
name = 'svr'
regressor = SVRRegressorRBF

model_params = {}
fit_params = {'C': [1, 5, 50, 100],
              'gamma': [0.05, 0.1, 0.5, 1],
              'cv': 5}
```

```python
for feature_combo in range(1, 15):
    model_name = '{}_{}'.format(name, feature_combo)
    
    print('\n')
    print('Model: {}'.format(model_name))
    print('\n')
    
    my_model =  MyModel(
                    model_name=model_name,
                    feature_combo=feature_combo,
                    regressor=regressor,
                    model_params=model_params,
                    fit_params=fit_params,
                    save_path=save_path,
                    data_path=data_path
                    )
    print('\n')
    
    my_model.fit_model()
    my_model.save()
    
    print('-'*25)
```
